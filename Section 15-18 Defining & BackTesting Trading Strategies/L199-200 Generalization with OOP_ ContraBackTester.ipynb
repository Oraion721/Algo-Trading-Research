```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

**Steps:**
1. Initialize the properties symbol,period,interval,etc
2. Add method `get_data()` for data retrieval and prepare the raw data.
3. Add the method `test_strategy()` with the parameter `window=`
4. Add the method `plot_result()`
5. Add the method `optimize_result()` no combination required, only one parameter
6. Add an appropriate `str` representation of `class ContraBackTester()` and `Docstrings`

```python
class ContraBackTester():
  def __init__(self,company):
    self.company=company
    self.get_data()

  # Mehtod: get_data()
  def get_data(self):
    data=yf.download(self.company,period="1mo",interval="5m")
    self.df=pd.DataFrame(data["Close"])
    self.df = self.df.ffill(axis=0)
    print(self.df)

  # Method: calculate_data()
  def calculate_data(self):
    for symbol in self.company:
      self.df[f"LR {symbol}"]=np.log(self.df[symbol]/self.df[symbol].shift(1))
      self.df[f"Position {symbol}"]=-np.sign(self.df[f"LR {symbol}"].rolling(window=3).mean())
      self.df[f"Strategy {symbol}"]=self.df[f"Position {symbol}"].shift(periods=1)*self.df[f"LR {symbol}"]
      self.df.dropna(inplace=True)
      print(self.df)
```

```python
tester=ContraBackTester(["BSE.NS","HINDCOPPER.NS","HINDZINC.NS","BTC-USD"])
```
