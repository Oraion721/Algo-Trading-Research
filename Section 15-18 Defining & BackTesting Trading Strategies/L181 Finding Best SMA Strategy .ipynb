# **Finding the best combination of SMA_S & SMA_L for SMA Strategy Backtesting**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
company="BEL.NS"
start_date="2023-01-01"
end_date="2026-01-06"
data=yf.download(company,start_date,end_date)
df=pd.DataFrame(data["Close"])
short_terms=7
long_terms=20
df
```

**Defining the function that will test the strategy for all combinations of SMA_S & SMA_L**

```python
def test_strategy(sma_combination):     # taking tuple as a argument
  df["SMA_S"]=df["BEL.NS"].rolling(window=sma_combination[0]).mean()
  df["SMA_L"]=df["BEL.NS"].rolling(window=sma_combination[1]).mean()
  df["Signal"]=np.where(df["SMA_S"]>df["SMA_L"],1,-1)
  df["LR"]=np.log(df["BEL.NS"]/df["BEL.NS"].shift(1))
  df["Strategy"]=df["Signal"].shift(1)*df["LR"]
  df.dropna(inplace=True)
  return np.exp(df["Strategy"].sum())
  # return df
```

```python
test_strategy((7,20))
```

**Creating Multiple combinations of SMA_S & SMA_L for test_strategy function**

```python
from itertools import product
combination=list(product(range(1,50,1),range(15,252,1)))
result=[]
for i in combination:
  result.append(test_strategy(i))
all_result=pd.DataFrame(data=combination,columns=["SMA_S","SMA_L"])
all_result["Performance"]=result
all_result.nlargest(5,"Performance")
```
