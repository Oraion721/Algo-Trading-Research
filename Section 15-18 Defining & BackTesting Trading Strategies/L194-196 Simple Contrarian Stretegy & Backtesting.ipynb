```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data=yf.download("BSE.NS",period="1mo",interval="5m")
df=pd.DataFrame(data["Close"])
df["LR"]=np.log(df["BSE.NS"]/df["BSE.NS"].shift(1))
df
```

**Defining Simple Contrarian Mean Strategy**

```python
df["Position"]=-np.sign(df["LR"].rolling(window=3).mean())
df
```

# **Vectorized Backtesting for Contrarian stretegy**

```python
df["strategy"]=df["Position"].shift(1)*df["LR"]
df.dropna()
```

```python
df[["LR","strategy"]].sum().apply(np.exp)
```

```python
df["Cumulative Return"]=df["LR"].cumsum().apply(np.exp)
df["Cumulative Strategy"]=np.exp(df["strategy"].cumsum())
df.dropna()
```

```python
df[["Cumulative Return","Cumulative Strategy"]].plot(figsize=(15,10),fontsize=12,title="BSE.NS | Window=3")
```

# **Changing Window Parameter**

```python
to_plot=["LR"]
for w in [1,2,3,5,7,10]:
  df[f"Position{w}"]=-np.sign(df["LR"].rolling(window=w).mean())
  df[f"Strategy{w}"]=df[f"Position{w}"].shift(1)*df["LR"]
  to_plot.append(f"Strategy{w}")
df.dropna()
```

```python
df[to_plot].plot(figsize=(15,8),fontsize=12,title=f"Returns at different windows")
plt.legend(fontsize=12)
plt.show()
```
