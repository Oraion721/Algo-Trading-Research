# **First Intuition on Iterative Backtesting**

**Importing libraries, data, and find the spread**

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
companies=["COCHINSHIP.NS","BDL.NS","FORTIS.NS"]
data=yf.download(companies,period="1y",interval="1d")
df=pd.DataFrame(data["Close"])
df_spread=pd.DataFrame(data["High"]-data["Low"])
sma_s=5;sma_l=20
for i in companies:
  df[f"Spread {i}"]=df_spread[i]
  df[f"LR {i}"]=np.log(df[i]/df[i].shift(1))
  # for bar in range(1,10):
  #   print(bar,i,df.index[bar].date(),df[i].iloc[bar],df[f"Spread {i}"].iloc[bar],sep="|")
  df[f"SMA_S {i}"]=df[i].rolling(sma_s).mean()
  df[f"SMA_L {i}"]=df[i].rolling(sma_l).mean()
df.dropna(inplace=True)
Position=0
df
```

```python
for i in companies:
  for j in range(len(df)):
    if df[f"SMA_S {i}"].iloc[j]>df[f"SMA_L {i}"].iloc[j]:
      if Position in [0,-1]:
        print(f"{df.index[j].date()} | {df[i]}: go Long | Price: {df[i].iloc[j]} | Spread: {df[f"Spread {i}"].iloc[j]}")
        Position=1
    elif df[f"SMA_S {i}"].iloc[j]<df[f"SMA_L {i}"].iloc[j]:
      if Position in [0,1]:
        print(f"{df.index[j].date()}: go Long| Price: {df[i].iloc[j]} | Spread: {df[f"Spread {i}"].iloc[j]}")
        Position=-1
```
